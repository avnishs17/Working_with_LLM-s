**Importing the Dependencies**

```python
import os

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
os.environ["HF_TOKEN"] = "your_huggingface_token"
```

```python
model_name = "google/gemma-2b-it"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
```

**Invoke the LLM**

```python
input_text = "Write a poem about Machine Learning"
```

```python
input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
```

```python
input_ids
```

```python
response = model.generate(**input_ids, max_new_tokens=512)
```

```python
response
```

```python
output = tokenizer.decode(response[0], skip_special_tokens=True)
print(output)
```

**Deploy LLM with Gradio as API**

```python
import json
import gradio as gr
```

```python
def get_response(input_data):
    input_data = json.loads(input_data)
    input_text = input_data["query"]
    input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
    response = model.generate(**input_ids, max_new_tokens=512)
    output = tokenizer.decode(response[0], skip_special_tokens=True)
    return {"output": output}
```

```python
demo = gr.Interface(
    fn=get_response,
    inputs="json",
    outputs="json"
)

demo.queue().launch(share=True, debug=True, show_error=True)
```
